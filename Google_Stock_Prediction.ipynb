```python
import pandas as pd
```

```python
file = '/content/GOOG.csv'
df = pd.read_csv(file)
```

```python
print(df.info())
```

```python
df.head(10)
```

```python
df['date'] = pd.to_datetime(df['Date'])
df.set_index(df['Date'],inplace=True)
```

```python
df.drop(columns=['Open','High','Low','Volume','Adj Close'],inplace=True)
```

```python
df.head(10)
```

```python
import plotly.graph_objects as go
```

```python
 trace1 = go.Scatter(
     x=df['Date'],
     y=df['Close'],
     mode='lines',
     name='Date'
)

 layout=go.Layout(
     title='Google Stock',
     xaxis={'title':'Date'},
     yaxis={'title':'Close'}
 )
 fig=go.Figure(data=[trace1],layout=layout)
 fig.show()
```

```python
Close_data=df['Close'].values
```

```python
Close_data
```

```python
#Converting array into 2-D arraay with one column
Close_data=Close_data.reshape(-1,1)
Close_data
```

```python
split_percentage=0.80

split=int(split_percentage * len(Close_data))
```

```python
split
```

```python
close_train = Close_data[:split]
close_test = Close_data[split:]
```

```python
print(len(close_train))
```

```python
date_train =df['Date'][:split]
date_test = df['Date'][split:]
```

```python
print(len(date_train))
```

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
look_back = 15
train_generator =TimeseriesGenerator(close_train,close_train,length=look_back,batch_size=20)
```

```python
print(train_generator)
```

```python
test_generator = TimeseriesGenerator(close_test,close_test,length=look_back,batch_size=1)
```

```python
model = Sequential()

model.add(
    LSTM(10,activation='relu',input_shape=(look_back,1))
    )
model.add(
    Dense(1)
)
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit_generator(train_generator,epochs=100)
```

```python
model.save('time.h5')
```

```python
prediction=model.predict_generator(test_generator)
```

```python
prediction
```

```python
prediction = prediction.reshape((-1))
close_train = close_train.reshape((-1))
close_test = close_test.reshape((-1))
```

```python
import matplotlib.pyplot as plt
plt.plot(close_train)
plt.plot(close_test)
plt.plot(prediction)
```

```python
 trace1 = go.Scatter(
    x=date_train,
    y=close_train,
    mode='lines',
    name='train data'
 )
 trace2 = go.Scatter(
     x=date_test,
     y=close_test,
     mode='lines',
     name='test data'
 )
 trace3 = go.Scatter(
     x=date_test,
     y=prediction,
     mode='lines',
     name='prediction'
 )
 layout = go.Layout(
     title = 'Google Stock',
     xaxis = {'title':'Date'},
     yaxis = {'title':'close'}
 )
 fig=go.Figure(data = (trace1,trace2,trace3),layout=layout)
 fig.show()
```
